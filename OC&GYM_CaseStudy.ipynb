## Define useful functions to support the next set of analysis

**Function #1:** To calculate overall churn

We define a function for calculating the overall churn:

$$churn\ rate = \Big(\frac{starting\ members + joining\ members - leaving\ members\ }{stating\ members}\Big)-1$$

```python
def get_churn():
    # Initialise 
    churn_data = {}
    no_of_members = 0

    # for each month
    for y in range(2009,2019):
        for m in range(1,13):

            # Build date, and get the underlying data required
            date = pd.datetime(y,m,1)
            data = subs_data_full.copy()

            # Compute joining members and leaving members
            joining_members = data[(data['Join Date'] == date)]['CustomerID'].count()
            leaving_members = data[(data['End Date'] == date)]['CustomerID'].count()

            # Calculate and record churn
            # Skip first round, to skip division by zero
            if no_of_members != 0:    
                churn_rate = ((no_of_members + joining_members - leaving_members) / no_of_members) - 1
                churn_data[date] = [no_of_members, joining_members, leaving_members, churn_rate]
            
            # Update number of members
            no_of_members += joining_members - leaving_members
    
    # Convert churn data into dataframe
    churn_data = pd.DataFrame.from_dict(churn_data, orient='index')
    churn_data.rename(columns={0:'starting_members', 1:'joining_members',
                               2:'leaving_members', 3:'churn_rate'}, inplace=True)
    return churn_data
```

Example to illustrate how this function works

**Function #2:** To calculate key stats given a feature

This function takes in a slice of the visitation data (original or enriched), together with a feature, and computes stats on membership size, revenue and churn.

```python
def get_stats_by_feature(subs_data_enriched, feature):
    # Initialise counter and dictionaries
    churn_data = []
    no_of_members = {}
    no_of_flex_members = {}
    no_of_standard_members = {}
    enriched_data = subs_data_enriched
     
    # for each element for the chosen feature
    for f in list(set(enriched_data[feature])):
        
        # Initialise counter
        no_of_members[f] = 0
        no_of_flex_members[f] = 0
        no_of_standard_members[f] = 0
        
        # Get underlying data                   
        data = enriched_data[(enriched_data[feature] == f)]
    
        # for each month
        for y in range(2009,2019):
            for m in range(1,13):
            
                # Build date
                date = pd.datetime(y,m,1)

                # Compute joining members and leaving members
                # New standard members
                no_of_new_std = data[(data['Join Date'] == date) &
                                   (data['Subscription Type'] == 'Standard')]['CustomerID'].count()
                # New flexible members
                no_of_new_flex = data[(data['Join Date'] == date) &
                                   (data['Subscription Type'] == 'Flexible')]['CustomerID'].count()
                joining_members = no_of_new_std + no_of_new_flex
                
                # Standard members leaving
                no_of_std_leaving = data[(data['End Date'] == date) &
                                   (data['Subscription Type'] == 'Standard')]['CustomerID'].count()
                # flexible members leaving
                no_of_flex_leaving = data[(data['End Date'] == date) &
                                   (data['Subscription Type'] == 'Flexible')]['CustomerID'].count() 
                leaving_members = no_of_std_leaving + no_of_flex_leaving
                
                # Compute new joiner fee collected
                joining_fee = joining_members * 5.00
                
                # Compute monthly fee collected
                if y <= 2015:
                    flex_fee = (no_of_flex_members[f] + no_of_new_flex - no_of_flex_leaving) * 25.99
                    std_fee = (no_of_standard_members[f] + no_of_new_std - no_of_std_leaving) * 20.99
                else:
                    flex_fee = (no_of_flex_members[f] + no_of_flex_leaving) * 30.99
                    std_fee = (no_of_standard_members[f] + no_of_new_std - no_of_std_leaving) * 26.99
                
                # Calculate total revenue
                total_revenue = joining_fee + flex_fee + std_fee

                # Calculate and record churn
                # Skip first round, to skip division by zero
                if no_of_members[f] != 0:    
                    churn_rate = ((no_of_members[f] + joining_members - leaving_members) / no_of_members[f]) - 1
                    churn_data.append([date, f, no_of_members[f], no_of_new_std, no_of_new_flex, joining_members, 
                                       leaving_members, (no_of_members[f] + joining_members - leaving_members), 
                                       churn_rate, joining_fee, std_fee, flex_fee, total_revenue])

                # Update number of members
                no_of_members[f] += joining_members - leaving_members
                no_of_standard_members[f] += no_of_new_std - no_of_std_leaving
                no_of_flex_members[f] += no_of_new_flex - no_of_flex_leaving
    
    # Convert churn data into dataframe
    churn_data = pd.DataFrame(churn_data)
    churn_data.rename(columns={0:'date', 1:feature, 2:'starting_members', 3:'new_std', 4:'new_flex',
                               5:'joining_members',6:'leaving_members', 7:'ending_members', 
                               8:'churn_rate', 9:'joining_fee', 10:'standard_fee', 11:'flex_fee',
                               12:'total_revenue'}, inplace=True)
    
    return churn_data
```

### Part 2: Visit dependent analysis

**Approach:**   
We use the visit dataset to build two new features to assist analysis:
- **Peak Time member ('PeakOrNotBoolean'):** to identify members who attend the gym primarily during peak hours
- **Frequent gym goers ('frequent_goer'):** member who attend the gym more than 12 times a months (or 3 times a week)
  
The logic for 'PeakOrNotBoolean' is: 
- For each month, if more than 50% of the visits happen at peak time, we label the month as 1, otherwise -1
- We then sum up this number for all months (for a member), and if the result is greater than 0, the member is classified as a peak time member, otherwise not a peak time member
  
**Insight and Recommendations:**
1. Peak time members seems to have plateau-ed and declining in 2018, this might be related to over-crowdedness at peak times.  We recommend a deeper analysis and survey of peak time members to understand whether that's the case.  Furthermore, the membership structure could be ammended to provide peak-time and non-peak time variants of both Flexible and Standard memberships, with the peak-time variant at a premium, and the non-peak variant at a discount (to current levels), this would bring in more revenue (given Peak Time members bring in more revenue), provide more control on capacity and also cater for potential overcowded-ness during peak hours.  Non-peak users also benefit from cheaper membership, and that might entice a new non-peak clientelle to join the gym.
2. We observe through the churn-rate that non-peak time members were more sensitive to the price change in 2016
3. Similarly, we find that non-frequent gym goers are more sensitive to price change

**Step 1:** Pre-process data, and update date type

```python
data = visit_data_full.copy()
data.reset_index(drop=True,inplace=True)
data.loc[:,'visitdate'] = pd.to_datetime(data.loc[:,'visitdate'], format = '%Y-%m-%d')

# Inspect
data.head(2)
```

**Step 2:** For each member, and for each month, compute the total number of visits and the number of peak time visits

```python
data_count_visit = data.groupby(['CustomerID', data.visitdate.dt.year, data.visitdate.dt.month])\
                        .agg({ 'visitdate': 'count', 'peakvisits': 'sum'})
data_count_visit.columns = ['VisitCount','PeakVisitCount']
data_count_visit.rename_axis(['CustomerID','Year','Month'],inplace=True)

# Inspect
data_count_visit.head(3)
```

**Step 3:** Calculate whether a member is a peak time user of not, as well as frequent gym goers (as at least 12 or more times a month)

```python
# Use logic describe above to tag each month (for each member) as peak or not
data_count_visit = data_count_visit.assign(PeakOrNot = lambda x: 2*data_count_visit.loc[:,'PeakVisitCount'] > data_count_visit.loc[:,'VisitCount'])
data_count_visit.replace({'PeakOrNot': False}, -1, inplace=True)

# Identify Frequent gym goers
data_count_visit.loc[:,'frequent_flag'] = data_count_visit.loc[:,'VisitCount'].apply(lambda x: 1 if x > 12 else -1)

# Find out if on aggregate (across all months) the use is a peak user or not, and a frequent gym goer or not
peak_analysis = data_count_visit.groupby('CustomerID').sum()
peak_analysis.loc[:,'PeakOrNotBoolean'] = peak_analysis.loc[:,'PeakOrNot'].apply(lambda x: True if x > 0 else False)
peak_analysis.loc[:,'frequent_goer'] = peak_analysis.loc[:,'frequent_flag'].apply(lambda x: True if x > 0 else False)
peak_analysis.head(5)
```

### Part 3: Macro Effect Analysis

### Part 3A:  Analysis of the effects of gym crowdedness (Utilisation)

**Hypothesis:**  
Members who are subscribed to crowded gyms should have a higher churn, as they get fed up from waiting for equipment and facilities, which leads them to leave.  

**Approach:**  
We classify each gym as either **Low**, **Normal** or **High** ultilisation based on its average daily utilisation compared to its maximum capacity.  We use the all-time maximum daily visit as a proxy for the max capacity for a given gym. We create an arbitrary set of global utilisation bands (Low range is 76.9% - 80.7% utilisation, Normal range is 80.7% - 84.4% ultilisation, and High range is 84.4% - 88.2% utilisation), based on the utilisation range across all the gyms, and use this to classify each gym.

**Observations and Insights:**  
1. We observe the same churn rates across all gym utilisations levels during the 2016 price change
2. We observe that members frequenting gyms classified as having **High** utilisation are on the wane; they are decreasing over the last 2 years of data.  Members of **Normal** utilisation gyms are flat, while members of **Low** utilisation gyms are on the rise.

**Recommendations:**  
1. We recommend reviewing down the subscription price for **High** utilisation gyms, to reverse the trend of decreasing membership observed over the past 2 years.
2. We recommend closely watching the subscription levels and revenue of the **Normal** utilisation gyms, as they have plateaued for the last 2 years, if the subscriptions numbers start to go down, consider reducing the subscription price to entice members to join/stay.
  

**Step 1:** Pre-process Data

```python
# Convert visitdate to a datetime format for processing
visit_data_full['visitdate'] = pd.to_datetime(visit_data_full['visitdate'])

# Take a subset of the data, from 2013 to 2018
visit_data = visit_data_full[(visit_data_full['visitdate'] > pd.datetime(2013, 1, 1))]

# Add the Club to each visit record
visit_data = visit_data.merge(subs_data_full[['CustomerID', 'Club']], left_on='CustomerID', 
                              right_on='CustomerID', how='inner')
```

**Step 2:** Compute Max Capacity for each Club

```python
# Get the total numnber of visit for each club on each day
max_cap_data = visit_data.groupby(['Club', 'visitdate']).agg({'CustomerID':'count'}).reset_index()

# For each club, find the max daily visit
max_cap = max_cap_data.groupby(['Club']).agg({'CustomerID':'max'})
max_cap.rename(columns={'CustomerID':'max_capacity'}, inplace=True)
```

**Step 3:** Compute Utilisation Rate for each Club

```python
# Get average daily visits for each club
util_rate_data = max_cap_data.groupby(['Club']).agg({'CustomerID':'mean'})
util_rate_data.rename(columns={'CustomerID':'mean_usage'}, inplace=True)

# Compute utilisation, i.e. average daily visits / max capacity
util_rate = util_rate_data.merge(max_cap, left_on='Club', right_on='Club', how='inner')
util_rate['util_rate'] = util_rate['mean_usage'] / util_rate['max_capacity']

# Assign a Low/Normal/High ultilisation category to each club
# First create some step changes
util_range_steps = (util_rate['util_rate'].max() - util_rate['util_rate'].min()) / 3
step_1, step_2 = util_rate['util_rate'].min() + util_range_steps, util_rate['util_rate'].min() + 2 * util_range_steps

# Report the ranges used
print ('Low range    :', round(util_rate['util_rate'].min(),3), '-', round(step_1,3))
print ('Normal range :', round(step_1,3), '-', round(step_2,3))
print ('High range   :', round(step_2,3), '-', round(util_rate['util_rate'].max(),3))

# Use steps to assign ultilisation category
util_rate['util_category'] = util_rate['util_rate'].apply(lambda x: 'High' if x >= step_2 
                                                          else ('Normal' if x >= step_1 else 'Low'))

# Inspect data (5 records)
util_rate.head()
```

**Step 4:** Analyse data across Utilisation Levels

```python
# Add new feature to data set
subs_data_enriched = subs_data_full.merge(util_rate[['util_category']], left_on='Club', right_on='Club')

analyse_feature(subs_data_enriched, 'util_category')
```

**Step 1:** Define a function to help analyse the joining and leaving trends

```python
def analyse_trend(data, year, month, title):
    
    # Get monthly number of members affected
    plot_data = data.groupby([year, month]).agg({'CustomerID':'count'}).reset_index()
    plot_data.rename(columns={'CustomerID':'members_month'}, inplace=True)

    # Get yearly number of members affected
    members_yearly = data.groupby([year]).agg({'CustomerID':'count'}).reset_index()
    members_yearly.rename(columns={'CustomerID':'members_year'}, inplace=True)

    # Calculate monthly number as a percentage of the yearly number
    plot_data = plot_data.merge(members_yearly, left_on=year, right_on=year)
    plot_data['monthly_pct'] = round(plot_data['members_month']/plot_data['members_year'], 3)
    
    # Plot trend
    plt.figure(figsize=(10,5))
    ax = sns.lineplot(x=year, y="monthly_pct", hue=month,
                      data=plot_data[(plot_data[year]>=2013)])
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.ylabel('Month as a % of Total Year')
    plt.title(title)
```

### Part 3D: Analysis of Membership Duration

**Hypothesis:**  
Better understanding what affects how long a member decides to keep his/her gym membership equips us with insight we can use to optimise our pricing and marketing strategies.

**Approach:**  
To carry out this analysis we first define membership duration, as the time between the subscription 'Join Date' and 'End Date'; for example if a member joins in Jan-18 and leaves in Apr-18, their membership duration is 3 months.  Similarly, if they join and leave in the same month, their membership duration is 1 month.  We then look at how the membership duration varies across all the different features (Subscription Type, Gender, Affluence, Age Bracket,Peak Time User, Frequent Goer, and year). 

**Insights and Recommendations:**
1. A large number of **Flexible members** leave within the first few months (Month 1: ~22%, Month 2: ~15% and Month 3: ~28%). We recommend the implementation of a "payment holiday" mechanism, so flexible members can pause their memberships and resume later, instead of having to cancel; this might reduce the month 1-2 dropout rate.
2. A large number of **Standard members** leave on Month 3 (~28%).  We recommend the implementation of an optional 6-month lock-in period (minimum stay) at a discounted price (e.g. 10%) for the first 6 months.
3. The **least affluent members (DE)** are more prone to leave early (within 3 month).  We recommend the extension of the payment holiday mechanism suggested above to less affluent members, so they don't feel pressured to have to cancel their membership, but instead can pause/resume.
4. The **<25 year-olds** are more prone to leave early (within 3 month).  We recommend the implementation of a discounted "young-people" rate, with a minimum lock-in (and discount) period of 6 months.
5. The **Peak Time** members are more prone to leave early (within 3 month).  We recommend a system to identify and engage with peak time members early on into their membership, to get feedback on their experience and resolve any concerns they might have (e.g. gym too crowded at peak time, class timetables...), to prevent early cancellations.
6. The **infrequent gym goers** are more prone to leave early (within 3 month). We recommend a system to identify and engage with infrequent gym goers early on into their membership, to get feedback on their experience and resolve any concerns they might have (e.g. don't like classes, don't know how to use machines...), to prevent early cancellations. 
7. Year 2018 has seen a material hike in early leavers, compared to the previous 5 years, which are broadly inline. We recommend carrying out a market analysis to understand whether there are exogeneous factors explaining the higher than usual drop-out rates (e.g. competitor pricing, disruptors like SoulCycle, subtler trends...).

### Part 3E - Analysis of Churn by duration of membership, using visitation data set

##### Rationale for separating effects of membership duration and  the effects of join month:

People who leave after joining the gym for a couple of months will have different reasons to those who leave after joining the gym for years, so the pricing strategy and the client relationship management should be different for these groups. We want to understand the pattern of churn rates at different membership duration (i.e. duration, see definition below).

We also want to understand if there are any trends in churn rates at different membership duration over the years. 

To separate these two factors, we divide the members into 118 cohorts by the calendar months that they joined the gym, and then track the churning pattern of each cohort, i.e. at which duration of membership they left the gym. 

Duration is defined as the number of *complete* months of membership, so if a person joins at the beginning of the month and leaves in the middle of that month, the complete month of membership is 0. 

For example, 364 people joined the gym in May 2009, 15 people left after 1 month of joining, 11 left after 2 months, 100 left after 3 months, and so on. Therefore for people who joined in May 2009, the churn rate at duration 3 is 100/364 = 27%. Similarly, 625 people joined in June 2009, 31 people left after 0 months of joining, 20 left after 1 month, 88 left after 2 months, and 139 left after 3 months, and so on. Therefore, for this cohort (people who joined in June 2009(, the churn rate for duration 3 is 139/625=22%. 

##### Assumptions

The visitation data does not have a Start date or End date as in the case of subscription data; consequently, we made the assumption that the first visit date is the join date and the last visit date is the end date for each member. 

The data ends on 27th Jan 2019, so if a person has visited the gym in Jan 2019, he/she would be classified as an "Active" member, and would not be counted as leavers. 

##### Key findings 

- The churn rates are the highest in the first 6 months of joining, in particular, the churn rates for duration 3 are the highest across all join months. 
- The churn rates for Jan 2016 are the highest across all durations, this is likely to be caused by the price increase.
- Price increased by 19% for both flexible and standard membership in Jan 2016, and this price change has increased churn rates at duration 1-6 by 7%-26% (absolute, not multiplicative changes). 
- The churn rates are fairly stable until late 2017/early2018. It seems that there are upward trends for churn rates at early durations (durations 0 -6) and a downward trend in churn rates at long durations (duration 12+). 

##### Recommended Actions:

- Offer deals and discounts for new joiners to keep these members, especially at duration 3.  Discount amount can be gradually reduced to nil pass 1 st anniversary.
- Investigate why people start to leave at early duration, perhaps conduct exit interviews. 
- 19% price increase caused churn rates to go up by as much as 26%, this means that the effect of the drop in the number of members might outweigh the effect of the increase in price. We should consider conducting a market survey if we want to increase price in future.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialise and train classifier
def decision_tree(X_train_val, y_train_val, X_test, X_train, y_train, X_val, y_val, y_test, seed = 100):
    DT_default = DecisionTreeClassifier(random_state = 100)
    DT_default.fit(X_train,y_train)

    # Predict output for Training and Validation data sets
    y_train_pred = DT_default.predict(X_train)
    y_val_pred = DT_default.predict(X_val)

    
    acc_train = accuracy_score(y_train,y_train_pred)
    acc_val = accuracy_score(y_val,y_val_pred)
    
    diff_acc = acc_train-acc_val
    
    DT_default.fit(X_train_val,y_train_val)
    y_test_pred = DT_default.predict(X_test)
    acc_test = accuracy_score(y_test, y_test_pred)
    
    print ('Training set accuracy  : ', acc_train)
    print ('Validation set accuracy: ', acc_val)
    print ('Test set accuracy:', acc_test )
    print('Diff of Acc:', diff_acc)
    return (diff_acc, acc_train, acc_val, acc_test)
```

```python
#Due to the size of the combined file we decide to shuffle the data set and then split into 10 parts
def split_10(df_full):
    shuffled = df_full.sample(frac=1)
    #Create new dataframe for each of the 10 divided data sets
    dfs = [pd.DataFrame(result) for result in np.array_split(shuffled,10)]
    return dfs
```

```python
def produce_dummy_and_seperate(X):
    X_filter = X.loc[:,['Club', 'Affluence', 'Gender', 'Subscription Type',
             'BirthYear', 'PeakOrNotBoolean','Joining_Age']]
    X_for_lm = pd.get_dummies(X_filter)
    
    #seperating
    X_new = X_for_lm.loc[:,['BirthYear', 'Club_Barking', 'Club_Barnet', 'Club_Bexley', 'Club_Brent',
   'Club_Bromley', 'Club_Camden', 'Club_Croydon', 'Club_Ealing',
   'Club_Enfield', 'Club_Greenwich', 'Club_Hackney', 'Club_Hammersmith',
   'Club_Haringey', 'Club_Harrow', 'Club_Havering', 'Club_Hillingdon',
   'Club_Hounslow', 'Club_Islington', 'Club_Kensington and Chelsea',
   'Club_Kingston upon Thames', 'Club_Lambeth', 'Club_Lewisham',
   'Club_Merton', 'Club_Newham', 'Club_Redbridge',
   'Club_Richmond upon Thames', 'Club_Southwark', 'Club_Sutton',
   'Club_Tower Hamlets', 'Club_Waltham Forest', 'Club_Wandsworth',
    'Affluence_A', 'Affluence_B', 'Affluence_C1',
   'Affluence_C2', 'Gender_Female',
   'Subscription Type_Flexible', 
    'PeakOrNotBoolean_True','Joining_Age']]
    y_new = X.loc[:,['duration']]
    
    return (X_new,y_new)
```

```python
def group_by_bins(y):
    #Predict first 1-3 months, then every 2 months
    y_with_bin_3_month = pd.cut(y.iloc[:,0],16,labels=False)

    #Predict first 1-6 months, then every 5 months
    y_with_bin_6_month = pd.cut(y.iloc[:,0],9,labels=False)
    return (y_with_bin_3_month,y_with_bin_6_month)
```

```python
from sklearn.neighbors import KNeighborsClassifier 

#KNN based on different period bins:
def KNN(X,y, starting=1, ending=400, interval=40):
    #Data splitting for KNN Model
    X_train_val, y_train_val, X_test, X_train, y_train, X_val, y_val, y_test = data_splitting(X,y)
    
    KNN_Acc_dict = {}
    # A dictionary to store the accuracy and ROC AUC Scores for each K-Nearest Neighbour classifier
    for k in range (starting,ending,interval):
        knn = KNeighborsClassifier(n_neighbors=k, p=2)
        knn.fit(X_train, y_train)
        pred_i = knn.predict(X_val)

        # Calculate Accuracy
        accuracy = accuracy_score(y_val, pred_i)
        print('When k=:',k,', the val accuracy is',accuracy)
        KNN_Acc_dict[k] = accuracy
    return KNN_Acc_dict
    
    
```

```python
#X, y = produce_dummy_and_seperate(splited_dataset)
def accuracy_summary(df):
    X, y = produce_dummy_and_seperate(df)
    y3,y6 = group_by_bins(y)

    D3 = KNN(X,y3,1,400,100)
    D6 = KNN(X,y6,1,400,100)
    return (D3,D6)
```

```python
def max_k(D):
    return max(zip(D.values(), D.keys()))
```
